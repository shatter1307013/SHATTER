# Cleaning Dataset for House-A

### Converts raw (per second and occupants'activity) dataset into minute and zone-wise occupancy dataset

```python
num_days = 30
activity_occupants_house_A = pd.DataFrame(columns = ['House', 'Day', 'Minute', 'Occupant-1 Activity', 'Occupant-2 Activity'])

for day in range(num_days):
    print('A', day)
    record = []
    # Imprtaing Datasets
    filename = 'raw/house-A/DAY-' + str(day + 1) + '.txt'
    dataframe =  pd.read_csv(str(filename), header=None, sep = ' ').iloc[:, -2:]    
    # Establishing Column Names to Make it Easier to Work with
    dataframe.columns = ['Occ-1','Occ-2']

    occ_1_activities = dataframe['Occ-1'].to_list()
    occ_2_activities = dataframe['Occ-2'].to_list()
    
    for i in range(0, len(occ_1_activities), 60):
        print('A', day + 1, int(i/60))
        occ_1_activity_id = occ_1_activities[i]
        occ_2_activity_id = occ_2_activities[i]
        
        record.append(["A", day + 1, int(i/60), occ_1_activity_id, occ_2_activity_id])
    
    activity_occupants_house_A = activity_occupants_house_A.append(pd.DataFrame(record, columns=activity_occupants_house_A.columns), ignore_index=True)
activity_occupants_house_A = activity_occupants_house_A.reset_index(drop = True)

activity_occupants_house_A.to_csv('cleaned/Cleaned-Dataframe_House-A.csv')
```
