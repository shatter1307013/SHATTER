# Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.ticker import MaxNLocator
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

# Plotting Function

```python
def plotting_bar(train_accuracy, test_accuracy, location):
    n_groups = len(train_accuracy)
    
    labels=["Training Accuracy","Testing Accuracy"]

    fig, ax = plt.subplots(figsize=(8, 5))
    index = np.arange(n_groups)
    bar_width = 0.3
    opacity = 0.80

    mat = np.zeros((2,n_groups))

    mat[0,:] = train_accuracy
    mat[1,:] = test_accuracy

    for ind in range (2):
        rects2 = plt.bar(index + ind * bar_width, mat[ind,:], bar_width,
        alpha=opacity,
        label=labels[ind])

    plt.xlabel('Number of Training Days',fontsize="24")
    plt.ylabel('Accuracy (%)',fontsize="24")
    #plt.title('Frequency of different sensors in the attack vectors',fontsize="15")
    plt.ylim([0, 100])
    plt.xticks(index + bar_width, ('10', '15', '20', '25'), fontsize="15")
    plt.yticks(fontsize="15")
    plt.legend(labels,loc='upper left',fontsize="20")
    plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))
    plt.savefig(location + '.pdf', bbox_inches='tight')
```

# Reading ADM Datasets

# Plotting ADM Results for House-A Occupant-1

```python
adm_accuracy_house_A_occ_1
train_accuracy = adm_accuracy_house_A_occ_1["Train Accuracy"].values
test_accuracy = adm_accuracy_house_A_occ_1["Test Accuracy"].values
test_accuracy
plotting_bar(train_accuracy, test_accuracy, "../../plots/ADM-House_A_Occ-1")
```
