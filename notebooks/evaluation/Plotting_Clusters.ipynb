# Importing Libraries

```python
import matplotlib.pyplot as plt
import json
```

# Plotting Function

```python
def plotting(list_cluster):
    #ax = plt.figure().add_subplot(projection='3d')
    from matplotlib.collections import PolyCollection
    fig= plt.figure(figsize=(8,5.5))
    ax = plt.axes(projection='3d')

    legendlist=['Outside', 'Bedroom', 'Livingroom', 'Kitchen', 'Bathroom']

    colors = ['#1B77BD', '#F87F00', '#369B00', '#D12D21', '#8E68C2']

    color_index = 0
    for i in range(len(list_cluster)):
        plt.plot(*zip(*list_cluster[i]["points"]), colors[color_index], alpha = 0.5)
        if i < len(list_cluster) - 1:
            if list_cluster[i]["zone_id"] != list_cluster[i + 1]["zone_id"]: 
                color_index += 1

    plt.legend(legendlist,loc='best',fontsize="10")

    plt.gca().get_legend().legendHandles[0].set_color(colors[0])
    plt.gca().get_legend().legendHandles[1].set_color(colors[1])
    plt.gca().get_legend().legendHandles[2].set_color(colors[2])
    plt.gca().get_legend().legendHandles[3].set_color(colors[3])
    plt.gca().get_legend().legendHandles[4].set_color(colors[4])

    plt.yticks(fontsize="15")
    plt.xticks(fontsize="15")
    ax.zaxis.set_tick_params(labelsize=15)

    ax.set_xlabel("Time of the Day (minute)",fontsize="15")
    ax.set_ylabel("Zone",fontsize="15")
    ax.set_zlabel("Duration of Stay (minute)",fontsize="15")


    ytick_labels = ['Z-0', 'Z-1', 'Z-2', 'Z-3', 'Z-4']
    plt.yticks(range(len(ytick_labels)), ytick_labels)
    #ax.set_yticks([0,1,2,3,4])
    plt.yticks(range(len(ytick_labels)), ytick_labels)
    
    ax.set_zlim(0,100)
    ax.azim = 20
    ax.dist = 7
    ax.elev = 20
```

# Plotting Cluster for House-A and Occupant-1

```python
with open('../../data/shatter/Clusters-House_A_Occ-1.txt', 'r') as f:
    list_cluster = json.loads(f.read())
plotting(list_cluster)
plt.savefig("../../plots/Clusters-House_A_Occ-1" + '.pdf', bbox_inches='tight')
```
